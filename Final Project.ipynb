```python
import pandas as pd

# Load dataset
df = pd.read_csv('sentiment140.csv', encoding='latin-1', header=0)
print(df.head())
```

```python
# Convert all values in the 'text' column to strings
df['text'] = df['text'].astype(str)
```

```python
import re

# Clean the 'text' column
def clean_text(text):
    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    # Remove mentions (like @username)
    text = re.sub(r'@\S+', '', text)
    # Remove special characters
    text = re.sub(r'[^A-Za-z\s]', '', text)
    # Convert to lowercase
    text = text.lower()
    return text

df['text'] = df['text'].apply(clean_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize TF-IDF Vectorizer
vectorizer = TfidfVectorizer(max_features=5000)

# Convert text to TF-IDF features
X = vectorizer.fit_transform(df['text'])
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the label encoder
label_encoder = LabelEncoder()

# Encode the sentiment column in your dataset
y = label_encoder.fit_transform(df['sentiment'])
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score

# ---------------------------
# 1. Load dataset
# ---------------------------
# df = pd.read_csv("yourfile.csv")  # you already loaded df

# ---------------------------
# 2. Encode the sentiment
# ---------------------------
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(df['sentiment'])

# ---------------------------
# 3. TF-IDF vectorization
# ---------------------------
vectorizer = TfidfVectorizer(max_features=5000)
X_tfidf = vectorizer.fit_transform(df['text'])  # important!

# ---------------------------
# 4. Train-test split
# ---------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf, y, test_size=0.2, random_state=42
)

# ---------------------------
# 5. Train model
# ---------------------------
model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)

# ---------------------------
# 6. Evaluate
# ---------------------------
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred, average='weighted'))
```

```python
# 8. SAVE MODEL + VECTORIZER + LABELS
# ---------------------------
joblib.dump(model, "sentiment_model.pkl")
joblib.dump(vectorizer, "tfidf_vectorizer.pkl")
joblib.dump(label_encoder, "label_encoder.pkl")

print("Model saved successfully!")
```
